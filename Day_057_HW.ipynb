# 作業
- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(X[:,0], X[:,1])
plt.show()
plt.close()
```

```python
# hierarchical clustering
from sklearn.cluster import AgglomerativeClustering
import matplotlib.cm as cm
methods = ['single','average','complete']

for method in methods:
  fig, ax = plt.subplots(figsize=(7, 7))
  hc = AgglomerativeClustering(n_clusters=3, linkage=method)
  predictions = hc.fit_predict(X)
  colors = cm.nipy_spectral(predictions.astype(float) / 3)
  ax.scatter(X[:,0], X[:,1], c=colors)
  ax.set_title(f'method = {method}')

plt.show()
plt.close()
```
