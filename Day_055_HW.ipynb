# 作業

- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
X = iris.data
y = iris.target

df_x = pd.DataFrame(data=X, columns=iris.feature_names)
df_x['target'] = y

df_x.tail()
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

estimators = [('k_means_4', KMeans(n_clusters=4)),
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_3_random', KMeans(n_clusters=3, init='random')),
              ('k_means_3_init', KMeans(n_clusters=3, n_init=3))
              ]
```

```python
outputs = {}
for name, km in estimators:
  km.fit(X)
  outputs[name] = km.labels_
```

```python
plt.scatter(df_x['sepal length (cm)'], df_x['sepal width (cm)'], c=df_x['target'])
plt.show()
plt.close()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10))

for i, (name, labels) in enumerate(outputs.items()):
  row = i // 2
  col = i % 2
  axs[row, col].scatter(df_x['sepal length (cm)'], df_x['sepal width (cm)'], c=labels)
  axs[row, col].set_title(name)

plt.show()
plt.close()
```
