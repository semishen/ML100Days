# Hierarchical clustering with 2D toy datasets

# [作業目標]
- 將階層式聚類套用在 2D 樣板資料上, 來觀察幾種不同參數的結果有何不同

```python
# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

'''
datasets.make_moons, datasets.make_blobs 
return ([n_samples, 2], [n_samples])
''' 
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15, 4))

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
ax1.scatter(noisy_moons[0][:,0],noisy_moons[0][:,1])
ax1.set_title('noisy_moons')

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
ax2.scatter(blobs[0][:,0],blobs[0][:,1])
ax2.set_title('blobs')

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2), None # convert to tuple
ax3.scatter(no_structure[0][:,0], no_structure[0][:,1])
ax3.set_title('no_structure')

plt.show()
plt.close()
```

```python
# 設定群集與繪圖參數
plot_num = 1
default_base = {'n_neighbors': 10, 'n_clusters': 3}
datasets = [(noisy_moons, {'n_clusters': 2}), (blobs, {}), (no_structure, {})]
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
# 執行各種樣板資料的繪圖迴圈
plt.figure(figsize=(10, 8))
for i_dataset, (dataset, algo_params) in enumerate(datasets):
  # 複製參數與標準化
  params = default_base.copy()
  params.update(algo_params)    
  X, y = dataset
  X = StandardScaler().fit_transform(X)

  # 設定三種不同參數的 Hierarchical clustering
  ward = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], linkage="ward")
  complete = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], linkage="complete")
  average = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], linkage="average")
  clustering_algorithms = (
      ('Average Linkage', average),
      ('Complete Linkage', complete),
      ('Ward Linkage', ward))

  # 繪製三種圖形
  for name, algorithm in clustering_algorithms:
    # t0 / t1 : 紀錄時間差
    t0 = time.time()
    algorithm.fit(X)
    t1 = time.time()
    if hasattr(algorithm, 'labels_'):
        y_pred = algorithm.labels_.astype(np.int)
    else:
        y_pred = algorithm.predict(X)

    plt.subplot(len(datasets), len(clustering_algorithms), plot_num)
    if i_dataset == 0:
        plt.title(name, size=18)

    colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                          '#f781bf', '#a65628', '#984ea3',
                                          '#999999', '#e41a1c', '#dede00']),
                                  int(max(y_pred) + 1))))
    plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y_pred])

    plt.xlim(-2.5, 2.5)
    plt.ylim(-2.5, 2.5)
    plt.xticks(())
    plt.yticks(())
    plt.text(.99, .01, ('%.2fs' % (t1 - t0)).lstrip('0'), transform=plt.gca().transAxes, size=15,
              horizontalalignment='right')
    plot_num += 1
plt.show()
```
