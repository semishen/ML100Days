```python
import pandas as pd
import numpy as np
```

```python
contries = pd.read_csv('data_csv.txt')
contries
```

```python
population = np.random.randint(10,100000, 249) # unit 10,000
contries['population'] = population
contries
```

```python
contries = contries.iloc[:,[0,2]]
contries.head(5)
```

```python
indices = np.random.choice(249,25)
partial = contries.iloc[indices, :]
partial
```

```python
data = {}
data['國家'] = partial['Name'].tolist()
data['人口'] = partial['population'].tolist()
data = pd.DataFrame(data)
data
```

```python
# max pupulation
max_pop = data['人口'].idxmax()
print("上述資料人口最多的國家是：{}".format(data.iloc[max_pop,0]))
```
